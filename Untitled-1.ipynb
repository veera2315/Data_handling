```python
import sqlite3

# Create a connection to a SQLite database
conn = sqlite3.connect('sales_data.db')

# Write the data frames to the database
adv_2021.to_sql('adv_2021', conn)
adv_2022.to_sql('adv_2022', conn)

# Merge the two data frames using SQL
merged_sales = pd.read_sql_query(
    "SELECT * FROM adv_2021 INNER JOIN adv_2022 ON adv_2021.'Hierarchy Level 3' = adv_2022.'Hierarchy Level 3'",
    conn
)
```

```python
# Merge the two data frames on Location
concat_2021_22=pd.concat([adv_2021,adv_2022],axis=0)
#merged_sales = pd.merge(adv_2021, adv_2022, on='Hierarchy Level 3', suffixes=('_2021', '_2022'))
concat_2021_22.shape
```

```python
sales_2021_by_level = adv_2021.groupby(['Hierarchy Level 3', 'Product category'])['Sales Value'].sum().reset_index()
sales_2022_by_level = adv_2022.groupby(['Hierarchy Level 3', 'Product category'])['Sales Value'].sum().reset_index()
merged_sales_by_level = concat_2021_22.groupby(['Hierarchy Level 3', 'Product category'])[['Sales Value', 'Sales Value']].sum().reset_index()
```

```python
# group the data by Product Brand and calculate the total sales value
grouped_2021 = adv_2021.groupby('Product category')['Sales Value'].sum().reset_index()
grouped_2022 = adv_2022.groupby('Product category')['Sales Value'].sum().reset_index()
# plot the comparison of sales in 2021 and 2022 by Product Brand
sns.set(style="whitegrid")
g = sns.catplot(x="Product category", y="Sales Value", data=grouped_2021, kind="bar", height=5, aspect=2)
g.fig.suptitle('Sales in 2021 by Product category')
#plt.xticks(rotation=90)
plt.show()
sns.set(style="whitegrid")
g = sns.catplot(x="Product category", y="Sales Value", data=grouped_2022, kind="bar", height=5, aspect=2)
g.fig.suptitle('Sales in 2022 by Product category')
#plt.xticks(rotation=90)
plt.show()

# group the data by Product Brand and calculate the total sales value
grouped_2021 = adv_2021.groupby('Product Brand')['Sales Value'].sum().reset_index()
grouped_2022 = adv_2022.groupby('Product Brand')['Sales Value'].sum().reset_index()
# plot the comparison of sales in 2021 and 2022 by Product Brand
sns.set(style="whitegrid")
g = sns.catplot(x="Product Brand", y="Sales Value", data=grouped_2021, kind="bar", height=5, aspect=2)
g.fig.suptitle('Sales in 2021 by Product Brand')
plt.xticks(rotation=90)
plt.show()
sns.set(style="whitegrid")
g = sns.catplot(x="Product Brand", y="Sales Value", data=grouped_2022, kind="bar", height=5, aspect=2)
g.fig.suptitle('Sales in 2022 by Product Brand')
plt.xticks(rotation=90)
plt.show()

# group the data by Product Name and calculate the total sales value
grouped_2021 = adv_2021.groupby('Product Name')['Sales Value'].sum().reset_index()
grouped_2022 = adv_2022.groupby('Product Name')['Sales Value'].sum().reset_index()
# plot the comparison of sales in 2021 and 2022 by Product Name
sns.set(style="whitegrid")
g = sns.catplot(x="Product Name", y="Sales Value", data=grouped_2021, kind="bar", height=5, aspect=3)
g.fig.suptitle('Sales in 2021 by Product Name')
plt.xticks(rotation=90)
plt.show()
sns.set(style="whitegrid")
g = sns.catplot(x="Product Name", y="Sales Value", data=grouped_2022, kind="bar", height=5, aspect=3)
g.fig.suptitle('Sales in 2022 by Product Name')
plt.xticks(rotation=90)
plt.show()
```

```python
# group the data by Product Brand and Hierarchy Level 3, and calculate the total sales value
grouped_2021 = adv_2021.groupby(['Product Brand', 'Hierarchy Level 3'])['Sales Value'].sum().reset_index()
grouped_2022 = adv_2022.groupby(['Product Brand', 'Hierarchy Level 3'])['Sales Value'].sum().reset_index()

# plot the comparison of sales in 2021 and 2022 by Product Brand and Hierarchy Level 3
sns.set(style="whitegrid")
g = sns.catplot(x="Product Brand", y="Sales Value", hue="Hierarchy Level 3", data=grouped_2021, kind="bar", height=5, aspect=2)
g.fig.suptitle('Sales in 2021 by Product Brand and Hierarchy Level 3')
plt.xticks(rotation=90)
plt.show()

sns.set(style="whitegrid")
g = sns.catplot(x="Product Brand", y="Sales Value", hue="Hierarchy Level 3", data=grouped_2022, kind="bar", height=5, aspect=2)
g.fig.suptitle('Sales in 2022 by Product Brand and Hierarchy Level 3')
plt.xticks(rotation=90)
plt.show()
```
